# **3) Cleaning Data**

**Cleaning customer_df**


---



```python
# Handling missing values in the gender column
customer_df.fillna(value="Prefer not to say", inplace=True)
print("Missing values after cleaning:\n", customer_df.isna().sum())
```

**How has company's revenue and sales performed in recent months?**


---



```python
# Creating a df that contains order counts and revenues (granularity: Month)
# Resampling (Adjusting the order_date column frequency from days to month)
# Resampling is primarily used for aggregating time-series data across different frequency.
# monthly_orders_df stores order counts and store's revenue for each month.
monthly_orders_df = all_df.resample(rule='M', on='order_date').agg({
    "order_id":'nunique',
    "total_price":'sum'
})

# Changing the string formatting of the order_date to Year-month.
monthly_orders_df.index = monthly_orders_df.index.strftime('%B')

# Reverting the index column from order_date to an integer column (default).
monthly_orders_df = monthly_orders_df.reset_index()

# Renaming column
monthly_orders_df.rename(columns={
    "order_id":"order_count",
    "total_price":"revenue"
}, inplace=True)
monthly_orders_df.head()
```

```python
# Visualizing the company's sales
plt.figure(figsize=(10,5))
plt.plot(
    monthly_orders_df['order_date'],
    monthly_orders_df['order_count'],
    marker='o'
)
plt.title("Numbers of orders each month")
plt.show()
```

```python
# Visualizing the company's revenue
plt.figure(figsize=(10,5))
plt.plot(
    monthly_orders_df['order_date'],
    monthly_orders_df['revenue'],
    marker='o'
)
plt.title("Revenue each month (AUD)")
plt.show()
```

**Which product sell the most and the least?**

```python
# Bar plot to visualize sales counts of products based on product type
plt.barh(
    y = all_df.groupby('product_type')['quantity_y'].sum().sort_values().index,
    width = all_df.groupby('product_type')['quantity_y'].sum().sort_values()
)
# Displaying the plot
plt.show()
```

**When did the customer last make a transaction?**

**How often has a customer made a purchase in recent months?**

**How much money have customers spent in recent months?**



---

To answer these three questions, we can perform RFM analysis. RFM analysis is a common method to segment customers using 3 parameters: recency, frequency, and monetary.


*   Recency -> When did the customer last make a transcation.
*   Frequency -> How often has a customer made a purchase.
*   Monetary -> How much money have customers spent.

With RFM analysis, it would help us to identify which customers are high value customers and which customers are low value customers.



```python
# Make a dedicated df for rfm analysis
# Contain the order count, the maximum order date, and sum purchases (AUD) of each customer.
rfm_df = all_df.groupby(by='customer_id', as_index=False).agg({
    'order_id':'nunique',
    'order_date':'max',
    'total_price':'sum'
})
# Getting the most recent order date
latest_date = all_df['order_date'].max()
# Calculating the distance between the most recent date and each order date
rfm_df['order_date'] =  rfm_df['order_date'].apply(lambda x: (latest_date - x).days)
rfm_df.rename(
    columns={
        'customer_id':'customer_id',
        'order_id':'frequency',
        'order_date':'recency',
        'total_price':'monetary'
    },
    inplace=True
)
rfm_df.set_index('customer_id', inplace=True)
rfm_df.head()
```

```python
# Test
def create_rfm_df(df):
    rfm_df = df.groupby(by='customer_id', as_index=False).agg({
        'order_id':'nunique',
        'order_date':'max',
        'total_price':'sum'
    })
    latest_date = df['order_date'].max()
    rfm_df['order_date'] = rfm_df['order_date'].apply(lambda x: (latest_date - x).days)
    rfm_df.rename(
        columns={
            'order_id':'frequency',
            'order_date':'recency',
            'total_price':'monetary'
        },
        inplace=True
    )
    return rfm_df

rfm_df = create_rfm_df(all_df)
rfm_df.sort_values(by='recency').head(5)
```

```python
# Plotting RFM Bar Plot
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(24,6))
colors = ["#72BCD4"] * 5

# 1st plot: Recency
sns.barplot(
    x=rfm_df['recency'].sort_values().head(5).index,
    y=rfm_df['recency'].sort_values().head(5),
    hue=rfm_df['recency'].sort_values().head(5).index,
    palette=colors,
    legend=None,
    ax=ax[0]
)
ax[0].set_ylabel('Days')
ax[0].set_xlabel(None)
ax[0].set_title("Recency", loc="center", fontsize=15)


# 2nd plot: Monetary
sns.barplot(
    x=rfm_df['monetary'].sort_values(ascending=False).head(5).index,
    y=rfm_df['monetary'].sort_values(ascending=False).head(5),
    hue=rfm_df['monetary'].sort_values(ascending=False).head(5).index,
    palette=colors,
    legend=None,
    ax=ax[1]
)
ax[1].set_ylabel('AUD')
ax[1].set_xlabel(None)
ax[1].set_title("Monetary", loc="center", fontsize=15)


# 3rd plot: Frequency
sns.barplot(
    x=rfm_df['frequency'].sort_values(ascending=False).head(5).index,
    y=rfm_df['frequency'].sort_values(ascending=False).head(5),
    hue=rfm_df['frequency'].sort_values(ascending=False).head(5).index,
    palette=colors,
    legend=None,
    ax=ax[2]
)
ax[2].set_ylabel('Times')
ax[2].set_xlabel(None)
ax[2].set_title("Frequency", loc="center", fontsize=15)

plt.suptitle("RFM Bar Plot (X-axis : customer_id)", fontsize=18)
plt.show()
```
