```python
dataset_path = '/kaggle/input/fit-body-dataset/train.csv'
dataset_train = pd.read_csv(dataset_path)

dataset_train.head(20)
```

```python
feature = dataset_train.columns

print(feature)
```

```python
# Mendefinisikan variabel dan menghitung jumlah nilai null untuk setiap kolom
null_counts = {
    'age': dataset_train['age'].isnull().sum(),
    'diet': dataset_train['diet'].isnull().sum(),
    'drinks': dataset_train['drinks'].isnull().sum(),
    'drugs': dataset_train['drugs'].isnull().sum(),
    'education': dataset_train['education'].isnull().sum(),
    'ethnicity': dataset_train['ethnicity'].isnull().sum(),
    'height': dataset_train['height'].isnull().sum(),
    'income': dataset_train['income'].isnull().sum(),
    'job': dataset_train['job'].isnull().sum(),
    'sex': dataset_train['sex'].isnull().sum(),
    'smokes': dataset_train['smokes'].isnull().sum()
}

# Mencetak jumlah nilai null untuk setiap kolom dengan label
print("Jumlah nilai null untuk setiap kolom:")
for column, null_count in null_counts.items():
    print(f"{column}: {null_count}")
```

```python
# Melihat Value dan menghitung berapa banyak value tersebut muncul pada kolom
null_counts = {
    'age': dataset_train.age.value_counts(dropna=False),
    'diet': dataset_train.diet.value_counts(dropna=False),
    'drinks': dataset_train.drinks.value_counts(dropna=False),
    'drugs': dataset_train.drugs.value_counts(dropna=False),
    'education': dataset_train.education.value_counts(dropna=False),
    'ethnicity': dataset_train.ethnicity.value_counts(dropna=False),
    'height': dataset_train.height.value_counts(dropna=False),
    'income': dataset_train.income.value_counts(dropna=False),
    'job': dataset_train.job.value_counts(dropna=False),
    'sex': dataset_train.sex.value_counts(dropna=False),
    'smokes': dataset_train.smokes.value_counts(dropna=False)
}

# Mencetak 
print("Value untuk setiap kolom:")
for column, null_count in null_counts.items():
    print("")
    print(f"{column}: {null_count}")
```

```python
# # Melakukan One-Hot Encoding menggunakan get_dummies()
# encoded_df = pd.get_dummies(dataset_train, columns=['age','diet','drinks','drugs','education','ethnicity','height','income','job','sex','smokes'])

# # Menampilkan hasil One-Hot Encoding
# print(encoded_df)

# encoded_df.head()

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(dataset_train[['age','diet','drinks','drugs','education','ethnicity','height','income','job','sex','smokes']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['age','diet','drinks','drugs','education','ethnicity','height','income','job','sex','smokes']))
print(encoded_df)
```

```python
y = dataset_train.fit_body

features_names = ['age','diet','drinks','drugs','education','ethnicity','height','income','job','sex','smokes']
X = encoded_df
```

```python
from sklearn.tree import DecisionTreeRegressor

#For model reproducibility, set a numeric value for random_state when specifying the model
fit_body_model = DecisionTreeRegressor(random_state=1)


# Fit the model
# Fit model
fit_body_model.fit(X, y)
```
