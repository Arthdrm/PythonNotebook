```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

# 1) Univariate Viz

**Bar Plot**


---



```python
# =============== Bar Plot : Vertical ========================== #
cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta',
          'Surakarta','Surabaya', 'Medan', 'Makassar')
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
df = pd.DataFrame({
    'Cities':cities,
    'Populations': populations
})

# Creating the bar plot
plt.bar(x=df['Cities'], height=df['Populations'])
# Configuring x-axis labels to be tilted to 45 degrees
plt.xticks(rotation=45)
# Displaying the plot
plt.show()
```

```python
# =============== Bar Plot : Horizontal ========================== #
plt.barh(y=df['Cities'], width=df['Populations'])
plt.xlabel("Populations (Millions)")
plt.title("Population of cities in indonesia")
plt.show()
```

```python
# =============== Bar Plot Seaborn ========================== #
# Seaborn is built on top of matplotlib. It provides richer options for data viz.
sns.barplot(y=df['Cities'], x=df['Populations'], orient='h', color='blue')
plt.xlabel("Populations (Millions)")
plt.title("Indonesian cities populations")
plt.show()
```

**Pie Chart**


---



```python
# =============== Pie chart original  ========================== #
flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.2, 0, 0, 0)

# Pie-chart parameter:
# x -> The data that want to be visualized (quantitative data).
# explode -> Determine space between wedges.
# labels -> Labels for each wedge (data categories).
# colors -> Colors that going to be used for the chart.
# autopct -> Stands for automatic percent. Filled with string formatting for labels.

plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    explode=explode,
    autopct='%.1f%%'
)
# The autopct value above will display percentage value one digit after decimal.

plt.show()
```

```python
# =============== Doughnut Pie Chart ========================== #
plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    wedgeprops={'width':0.4}
)
plt.show()
```

**Histogram**


---



```python
# Histogram parameter:
# x -> Data to be visualized.
# bins -> Number of bins
x = np.random.normal(15, 5, 250)
plt.hist(x=x, bins=15)
plt.show()
```

**Box Plot**


---



```python
x = np.random.normal(15, 5, 250)
sns.boxplot(x)
plt.show()
```

# 2) Bivariate Viz

**Scatter Plot**


---



```python
# =============== Plain Scatter Plot ========================== #
lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15,
                  9.96, 7.21, 10.04, 10.2, 11.06]
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93,
                132.93, 138.92, 145.98, 148.44, 152.81]
plt.scatter(
    x=lemon_diameter,
    y=lemon_weight
)
plt.xlabel("Lemon Diameter")
plt.ylabel("Lemon Weight")
plt.show()
```

```python
# =============== Scatter Plot With Regression ========================== #
# Legend:
# Blue line -> Correlation line
# Band -> Confidence level of the regression function (high confidence = small width)
sns.regplot(
    x=lemon_diameter,
    y=lemon_weight
)
plt.show()
```

**Line Chart (Often used with time-series data)**


---



```python
url = 'https://query1.finance.yahoo.com/v7/finance/download/BBCA.JK?period1=1644796800&period2=1676332800&interval=1d&events=history&includeAdjustedClose=true'
df = pd.read_csv(url)
df['Date'] = pd.to_datetime(df['Date'])

# Adjusting figure size (width x color)
plt.figure(figsize=(12,5))
# Plotting the line chart on the current figure
plt.plot(df['Date'], df['Close'], color='red')
# Adding a label to the x-axis
plt.xlabel('Date', size=15)
# Adding a label to the y-axis
plt.ylabel('Price', size=15)
plt.show()
```

**Clustered Bar Chart (Multiple Bar Chart)**


---




```python
penguins = sns.load_dataset("penguins")
sns.barplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="sex",
    errorbar=None
)
plt.show()
```
