```python
def calculate_entropy(df, target_col):
  individual_class_count = df[target_col].value_counts().values
  total_class_count = len(df)
  p = individual_class_count/total_class_count
  return sum(-p * np.log2(p))
```

```python
def calculate_threshold(df):
  threshold_all = dict()
  # Loop except the last column (which is the class column)
  for col in df.columns[:-1]:
    threshold_all[col] = dict()
    sorted_col = df[col].sort_values().reset_index(drop=True)
    for x in range(len(df) - 1):
      threshold = (sorted_col[x] + sorted_col[x+1])/2
      threshold_all[col][threshold] = {
        'entropy': None,
        'gain': None,
        'split_info': None,
        'gain_ratio': None          
      }
  return threshold_all
```

```python
def calculate_entropy_attribute(df, threshold, target_col):
  e_attr = copy.deepcopy(threshold)
  for attr_name, attr_val in e_attr.items():
    for threshold, threshold_val in attr_val.items():
      df_filtered_1 = df[df[attr_name] <= threshold]
      df_filtered_2 = df[df[attr_name] > threshold]        
      e_1 = calculate_entropy(df_filtered_1, target_col)
      e_2 = calculate_entropy(df_filtered_2, target_col)
      e_attr[attr_name][threshold]['entropy'] = np.array([e_1, e_2])
  return e_attr
```

```python
def calculate_gain_attribute(df, e_total, e_attr, target_col):
    gain_attr = copy.deepcopy(e_attr)
    len_total = len(df)
    for attr_name, attr_val in gain_attr.items():
        for threshold, threshold_val in attr_val.items():
            len_1 = len(df[df[attr_name] <= threshold])
            len_2 = len(df[df[attr_name] > threshold])  
            len_threshold = np.array([len_1, len_2])
            gain = e_total - sum(len_threshold/len_total * threshold_val['entropy'])
            gain_attr[attr_name][threshold]['gain'] = gain
    return gain_attr  
```

```python
def calculate_split_info(df, gain_attr, target_col):
    split_info_attr = copy.deepcopy(gain_attr)
    len_total = len(df)    
    for attr_name, attr_val in split_info_attr.items():
        for threshold, threshold_val in attr_val.items():
            len_1 = len(df[df[attr_name] <= threshold])
            len_2 = len(df[df[attr_name] > threshold])  
            len_threshold = np.array([len_1, len_2])
            term = len_threshold/len_total
            term[term == 0] = 1 # Boolean mask to convert 0 to 1.
            split_info = sum(-term * np.log2(term))
            split_info_attr[attr_name][threshold]['split_info'] = split_info
    return split_info_attr      
```

```python
def calculate_gain_ratio(split_info_attr):
    gain_ratio_attr = copy.deepcopy(split_info_attr)
    for attr_name, attr_val in gain_ratio_attr.items():
        for threshold, threshold_val in attr_val.items():
            if threshold_val['split_info'] == 0:
                gain_ratio = 0
            else:
                gain_ratio = threshold_val['gain']/threshold_val['split_info']
            gain_ratio_attr[attr_name][threshold]['gain_ratio'] = gain_ratio
    return gain_ratio_attr
```

```python
def get_highest_gain_ratio(gain_ratio_attr):
    highest_ratio_value = float(-inf)
    highest_ratio_threshold = 0
    highest_ratio_attr_name = ""
    for attr_name, attr_val in gain_ratio_attr.items():
        for threshold, threshold_val in attr_val.items():
            if threshold_val['gain_ratio'] > highest_ratio_value:
                highest_ratio_value = threshold_val['gain_ratio']
                highest_ratio_threshold = threshold
                highest_ratio_attr_name = attr_name
    return [highest_ratio_attr_name, highest_ratio_threshold]
```

```python
def find_decision_node(df, class_col):
    # Calculate the entropy of the entire set
    e_total = calculate_entropy(df, class_col)
    # Perform discritization to numerical features
    threshold = calculate_threshold(df)
    # Calculate the entropy of each attribute (for each threshold)
    e_attr = calculate_entropy_attribute(df, threshold, class_col)
    # Calculate the gain of each attribute compared to the entire entropy
    gain_attr = calculate_gain_attribute(df, e_total, e_attr, class_col)
    # Calculate the SplitInfo of each attribute
    split_info_attr = calculate_split_info(df, gain_attr, class_col)
    # Calculate the GainRatio of each attribute
    gain_ratio_attr = calculate_gain_ratio(split_info_attr)
    # Get the attribute (with its threshold) that has the highest information gain
    highest_gain = get_highest_gain_ratio(gain_ratio_attr)
    return highest_gain
```

```python
# Plan for the rule set (The rule must be added during tree generation)
# Example
rule_set = {
    "11":{
        'attribute': 'PetalLengthCm',
        'threshold': 1.9,
        'class': 'Iris-setosa',
        'sign': '<=',
    },
    "12":{
        'attribute': 'PetalLengthCm',
        'threshold': 1.9,
        'class': None,
        'sign': '>',
    },
    "121":{
        'attribute': 'PetalLengthCm',
        'threshold': 4.9,
        'class': None,
        'sign': '<='
    },
    "1211":{
        'attribute': 'PetalWidthCm',
        'threshold': 1.6,
        'class': 'Iris-versicolor',
        'sign': '<='        
    }
}
```

```python
# Sample 5 rows out of df_train (that has either iris-setosa & iris-versicolor as a class)
df_2 = df_train[8:10]
df_2.insert(5, 'Prediction', ['?'] * len(df_2))
df_2
```

```python
def apply_rules(df, rule_set, target_col):
    df_predict = df.copy()
    attribute = ""
    threshold = 0
    label = ""
    sign = ""
    leaf_nodes = [key for key, value in rule_set.items() if value['class'] != None]
    for node in leaf_nodes:
        label = rule_set[node]['class']        
        if len(node) > 2:
            # Case when a leaf node is not direct child of the root node (where backtracking is needed)            
            node_chain = [node[:2+x] for x in range(len(node[1:]))]
            query_str = ""
            for x in node_chain:
                attribute = rule_set[x]['attribute']
                threshold = rule_set[x]['threshold']
                sign = rule_set[x]['sign']
                query_str += "{} {} {} & ".format(attribute, sign, threshold)
            query_str = query_str.rstrip('& ') # Removing the trailing and operator
            df_predict.loc[df_predict.query(query_str).index, target_col] = label
        else:
            # Case when a leaf node is a direct child of the root node.            
            attribute = rule_set[node]['attribute']
            threshold = rule_set[node]['threshold']
            sign = rule_set[node]['sign']
            query_str = "{} {} {}".format(attribute, sign, threshold)
            df_predict.loc[df_predict.query(query_str).index, target_col] = label
    return df_predict
```

```python
df_2_predict = apply_rules(df_2, rule_set, 'Prediction')
df_2_predict
```

Use df.query() to dynamically create AND statements. The flow of applying AND statement: Leaf node -> Decision Node. For example, to apply the rule 
at 121 node, first you must apply the rule at the node 12, then you apply the rule at node 121. This can be done by iterating the key (string) of the rule set dictionary. Like this key[:2] -> key[:3]. 

To strip the additional AND operator at the end of a string query, we can use rstrip.

Also remember to assign value with the syntax below: 
- df.loc[condition, col] = new_val

Rather than chaining index like this: 
- df[boolean_indexing][col] = new_val

Because it will prompt the SetttingWithCopyWarning.
